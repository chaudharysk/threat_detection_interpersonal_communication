```python
import pandas as pd
import os
```

```python
def text_preprocess(file,para_lines):
    with open(file) as f:
        lines = f.readlines()
        # removing blank lines
        lines = [x for x in lines if x!='\n']
        #removing /n from everywhere
        lines = [s.replace("\n", " ") for s in lines]
        # removing all /n from lines
        #print(pd.DataFrame(lines))   
    concatenated_list=[]
    for i in range(0, len(lines)-1,para_lines):
        concatenated_string = "".join(lines[i:i+para_lines])
        concatenated_string = concatenated_string+' ,'
        concatenated_list.append(concatenated_string)
    print(pd.DataFrame(concatenated_list)) 
    return concatenated_list
```

```python
dir_list = os.listdir('Data/Sub Data')
for file_path in dir_list:
    k = text_preprocess('Data/Sub Data'+'/'+file_path,20)
    with open('final.csv', 'a') as f:
        for x in k:
            f.write(x)
            f.write('\n')
```
